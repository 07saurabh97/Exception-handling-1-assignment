Q1. Explain why we have to use the Exception class while creating a Custom Exception.

When creating a custom exception in a programming language, it is common to extend the built-in Exception class or a subclass of it. Here's why using the Exception class is beneficial:

Inheritance: By extending the Exception class, the custom exception inherits all the properties and behaviors of the base Exception class. This includes methods like getMessage(), printStackTrace(), and toString(), which can be used to provide useful information about the exception when it is caught and handled.

Compatibility: Using the Exception class ensures that the custom exception is compatible with the existing exception handling mechanisms in the language. It allows the custom exception to be caught and handled using the same try-catch blocks that are used for built-in exceptions.

Consistency: By following the convention of extending the Exception class, the custom exception becomes consistent with other exceptions in the language. This makes the code more readable and maintainable, as developers familiar with the language can easily understand and handle the custom exception.

Documentation: The Exception class and its subclasses often come with well-documented and well-understood behavior. By extending the Exception class, the custom exception benefits from this documentation, making it easier for other developers to understand and use the custom exception in their code.

Exception Hierarchy: The Exception class is typically part of an exception hierarchy, with different subclasses representing different types of exceptions. By extending the Exception class, the custom exception can be placed in the appropriate position in the hierarchy, allowing for more specific exception handling based on the type of exception.

The ArithmeticError class in Python is the base class for all errors that occur during numeric calculations.

1. ZeroDivisionError- This error occurs when a number is divided by zero. It is raised when attempting to perform division or modulo operation with zero as the divisor.


```python
try:
    result = 15 / 0
except ZeroDivisionError as i:
    print("Error:", i)
```

2. OverFlowError- This error occurs when the result of an arithmetic operation exceeds the maximum representable value for a numeric type. It is raised when attempting to perform an operation that results in a value that is too large to be represented.

```python
try:
    result = 20 ** 1000
except OverflowError as i:
    print("Error", i)
```

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

The LookupError class in Python is the base class for exceptions that are raised when a key or index used on a mapping or sequence is invalid. 


```python
my_dict = {'name': 'Saurabh', 'age': 26}

try:
    print(my_dict['city'])
except KeyError as e:
    print("Error:", e)
```

Q5. Explain ImportError. What is ModuleNotFoundError?

ImportError is an exception that is raised when an import statement encounters difficulties in successfully importing a specified module. It is typically caused by an invalid or incorrect path to the module

ModuleNotFoundError occurs when Python is unable to find and import a specified module. This error is raised when you try to access or use a module that cannot be found in the Python environment.
